## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.


# Convert date components to strings and fill NaN values with empty strings
df['arrival_date_day_of_month'] = df['arrival_date_day_of_month'].fillna('').astype(str)
df['arrival_date_month'] = df['arrival_date_month'].fillna('').astype(str)
df['arrival_date_year'] = df['arrival_date_year'].fillna('').astype(str)

# Combine date components into a single date string column
df['arrival_date'] = df['arrival_date_year'] + '-' + df['arrival_date_month'] + '-' + df['arrival_date_day_of_month']

# Convert the combined date string column to datetime
df['arrival_date'] = pd.to_datetime(df['arrival_date'], errors='coerce')

# Drop unnecessary columns
df.drop(['arrival_date_year', 'arrival_date_month', 'arrival_date_week_number', 'arrival_date_day_of_month', 'reservation_status_date'], axis=1, inplace=True)

# Display the modified DataFrame
print(df.head())

````

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code

# Select numerical columns for correlation analysis
numerical_cols = df.select_dtypes(include='number')

# Compute the correlation matrix
corr_matrix = numerical_cols.corr()

# Plot the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
````

##### 1. Why did you pick the specific chart?

Answer Here.

Chose a correlation heatmap because it effectively visualizes the relationships between numerical variables in the dataset.

**Comprehensive Insight**: A correlation heatmap allows for a comprehensive view of the relationships between all pairs of numerical variables in the dataset. This enables us to identify patterns of association or dependency between variables, providing valuable insights into the underlying structure of the data.

**Visual Representation**: The heatmap provides a visual representation of correlation coefficients, making it easy to interpret and identify significant relationships between variables. Color gradients highlight the strength and direction of correlations, allowing for quick identification of patterns.

**Identification of Multicollinearity**: The heatmap helps identify multicollinearity, which occurs when two or more variables are highly correlated with each other. Detecting multicollinearity is crucial for regression analysis and predictive modeling, as it can affect model performance and interpretation of results.

Overall, the correlation heatmap is a powerful tool for understanding the relationships between numerical variables in the dataset and identifying potential patterns or dependencies that may influence business decisions or analytical insights.

##### 2. What is/are the insight(s) found from the chart?

Answer Here

**Positive Correlation**: Positive correlation coefficients (values closer to 1) indicate that as one variable increases, the other variable also tends to increase. For example, there may be a positive correlation between variables such as the number of adults and the total booking amount, suggesting that larger group bookings lead to higher revenue.

**Negative Correlation**: Negative correlation coefficients (values closer to -1) suggest that as one variable increases, the other variable tends to decrease. For instance, there may be a negative correlation between the length of stay and the average daily rate (ADR), indicating that longer stays are associated with lower daily rates.

**No Correlation**: Correlation coefficients close to 0 suggest little to no linear relationship between variables. For example, there may be no significant correlation between the number of children and the number of available parking spaces, indicating that these variables are independent of each other.

**Multicollinearity**: High correlation coefficients between pairs of variables may indicate multicollinearity, which can affect the reliability of regression models. Identifying multicollinearity allows for adjustments in model building to ensure accurate and interpretable results.

Overall, the correlation heatmap provides insights into the strength and direction of relationships between numerical variables in the dataset, helping to identify potential patterns, dependencies, and areas for further analysis or modeling.
